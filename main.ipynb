```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('s0064-sit-stand-balance.csv', header=None, names=['Value'])
```

## Cek Missing Value

```python
missing_values = df.isnull().sum()

total_missing = missing_values.sum()
print("\nTotal Missing Values:", total_missing)
```

## 1. Buatkan visualisasi didalam python

```python
print(df.head())

plt.figure(figsize=(20, 8))
plt.plot(df.index, df['Value'], label='Value')

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('s0064-sit-stand-balance')

plt.legend()
plt.show()
```

## 2. Silahkan teman teman deteksi/tandai/check mana yang menjadi puncaknya dan di titik berapa saja. Ada berapa titik puncaknya

```python
#membuat fungsi untuk mendeteksi puncak
def deteksi_puncak(data, threshold=0):
    puncak = []

    for i in range(1, len(data) - 1):
        if data[i] > data[i - 1] and data[i] > data[i + 1] and data[i] > threshold:
            puncak.append(i)

    return puncak

indeks_puncak = deteksi_puncak(df['Value'])

print("Data Puncak:")
for i, indek_puncak in enumerate(indeks_puncak, start=1):
    print(f"Puncak {i}: Index = {indek_puncak}, Nilai = {df['Value'].iloc[indek_puncak]}")


total_puncak = len(indeks_puncak)
print(f"\nJumlah Total Puncak: {total_puncak}")
```

## 3. Tandai puncak tersebut dengan scatter plot

```python
# Menghapus data yang bukan titik puncak
df_puncak = df.iloc[indeks_puncak]

plt.figure(figsize=(20,8))
plt.plot(df.index, df['Value'], label='Data Mentah', alpha=0.5)

# Memvisualisasi data titik puncak saja 
plt.scatter(df_puncak.index, df_puncak['Value'], color='red', label='Puncak')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Deteksi Titik Puncak')
plt.legend()
plt.show()

print('\nTampilan data yang di titik puncak saja')

# Memvisualisasi data titik puncak saja dan membuat data mentahnya tidak di tampilkan
plt.figure(figsize=(20, 8))
plt.scatter(df_puncak.index, df_puncak['Value'], color='red', label='Puncak')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Deteksi Titik Puncak')

# Menetapkan nilai sumbu y awal supaya datanya bisa terlihat ada ditengah
plt.ylim(-30000, df_puncak['Value'].max()) 
plt.legend()
plt.show()
```

## 4. Hitung jarak dari titik yang di tandai dengan titik setelahnya

```python
# Menghitung jarak antar titik puncak
jarak_puncak = np.diff(indeks_puncak)

print("Jarak Antar Titik Puncak:")
for i, jarak in enumerate(jarak_puncak, start=1):
    print(f"Jarak Puncak {i} dengan {i+1} = {jarak}")
```

## 5. Hasil jarak dari titik yang telah ditandai diplotkan menggunakan line chart

```python
# Memvisualisasikan jarak antar titik puncak dengan line chart
plt.figure(figsize=(20, 8))
plt.plot(range(1, len(jarak_puncak) + 1), jarak_puncak, marker='o', linestyle='-')

plt.xlabel('Nomor Indeks Puncak')
plt.ylabel('Jarak Antar Puncak')
plt.title('Grafik Line Chart Jarak Antar Puncak')
plt.show()
```
